```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
sales_by_product = df.groupby("Product")["Sales"].sum()
sales_by_product
```

```python
sales_by_product.plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
